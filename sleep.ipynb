```python
import pandas 
data = pandas.read_csv("sleep.csv")
data
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Sleep Disorder'].fillna('None', inplace = True)
```

```python
data.drop_duplicates()
data.head(20)
```

```python
data['BMI Category'].value_counts()
```

```python
data['BMI Category'].replace({
    'Normal Weight':'Normal'
},inplace=True)
data.head(20)
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns   
plt.style.available 
```

```python
plt.style.use('ggplot')
```

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Gender', data=data)
plt.title('Bar Plot of Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Workers')
for p in ax.patches:
  txt = str(p.get_height().round(1)) + ' '
  txt_x = p.get_x()
  txt_y = p.get_height()
  ax.text(txt_x,txt_y,txt)
plt.show()
```

```python
bmi_counts =data['BMI Category'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(bmi_counts, labels=bmi_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of BMI Category')
plt.show()
```

```python
import numpy
ax = sns.barplot(x = 'Gender', y = 'Quality of Sleep', data = data,palette = 'magma')
for p in ax.patches:
  txt = str(p.get_height().round(1)) + ' '
  txt_x = p.get_x()
  txt_y = p.get_height()
  ax.text(txt_x,txt_y,txt)
```

```python
bmi_counts =data['Sleep Disorder'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(bmi_counts, labels=bmi_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart of Sleep Disorder')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Sleep Duration'], bins=20, kde=True)
plt.title('Histogram of Sleep Duration')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data['Stress Level'], shade=True)
plt.title('Density Plot of Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Density')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Sleep Duration', data=data)
plt.title('Box Plot of Sleep Duration by Gender')
plt.xlabel('Gender')
plt.ylabel('Sleep Duration (hours)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Quality of Sleep', y='Physical Activity Level', hue='Gender', data=data)
plt.title('Line Plot of Quality of Sleep and Physical Activity Level by Gender')
plt.xlabel('Quality of Sleep (hours)')
plt.ylabel('Physical Activity Level')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='BMI Category', y='Sleep Duration', data=data)
plt.title('Bar Plot of Average Sleep Duration by BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Average Sleep Duration (hours)')
for p in ax.patches:
  txt = str(p.get_height().round(1)) + ' '
  txt_x = p.get_x()
  txt_y = p.get_height()
  ax.text(txt_x,txt_y,txt)
plt.show()
```

```python
ax = sns.FacetGrid(data, col="Gender", row="BMI Category", margin_titles=True)
ax.map(plt.hist, "Sleep Duration", bins=20)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Sleep Duration', y='Heart Rate', hue='Gender', data=data)
plt.title('Line Plot of Sleep Duration and Heart Rate by Gender')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Heart Rate')
plt.show()
```

```python
average_heart_rate_by_stress = data.groupby('Stress Level')['Heart Rate'].mean().reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x='Stress Level', y='Heart Rate', data=average_heart_rate_by_stress)
plt.title('Average Heart Rate by Stress Level Categories')
plt.xlabel('Stress Level Category')
plt.ylabel('Average Heart Rate')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(data=data, x='BMI Category', hue='Sleep Disorder', multiple='stack')
plt.title('Stacked Bar Plot of Sleep Disorder by BMI Category and Gender')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.show()
```

```python
average_stress_level_by_occupation = data.groupby('Occupation')['Stress Level'].mean().reset_index()

plt.figure(figsize=(14, 8))
sns.barplot(x='Occupation', y='Stress Level', data=average_stress_level_by_occupation)
plt.title('Average Stress Level by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Average Stress Level')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='BMI Category', y='Daily Steps', hue='Gender', data=data)
plt.title('Bar Plot of Average Daily Steps by BMI Category and Gender')
plt.xlabel('BMI Category')
plt.ylabel('Average Daily Steps')
plt.show()
```

```python
average_quality_of_sleep_by_disorder = data.groupby('Sleep Disorder')['Quality of Sleep'].mean().reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x='Sleep Disorder', y='Quality of Sleep', data=average_quality_of_sleep_by_disorder)
plt.title('Average Quality of Sleep by Sleep Disorder')
plt.xlabel('Sleep Disorder')
plt.ylabel('Average Quality of Sleep')
plt.show()
```

```python

plt.figure(figsize=(12, 8))
ax = sns.countplot(x='Occupation', data=data)
plt.title('Count Plot of Occupation')
plt.xlabel('Occupation')
plt.ylabel('Number of Workers')
for p in ax.patches:
  txt = str(p.get_height().round(1)) + ' '
  txt_x = p.get_x()
  txt_y = p.get_height()
  ax.text(txt_x,txt_y,txt)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Gender', y='Heart Rate', data=data)
plt.title('Violin Plot of Heart Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Heart Rate')
plt.show()
```

```python
average_heart_rate_by_bmi = data.groupby('BMI Category')['Heart Rate'].mean().reset_index()

plt.figure(figsize=(12, 8))
ax = sns.barplot(x='BMI Category', y='Heart Rate', data=average_heart_rate_by_bmi)
plt.title('Average Heart Rate by BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Average Heart Rate')
for p in ax.patches:
  txt = str(p.get_height().round(1)) + ' '
  txt_x = p.get_x()
  txt_y = p.get_height()
  ax.text(txt_x,txt_y,txt)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='BMI Category', y='Sleep Duration', data=data)
plt.title('Box Plot of Sleep Duration by BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Sleep Duration (hours)')
plt.show()
```
